### Sprint 8: AnÃ¡lisis sintÃ¡ctico de HTML

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

URL = 'https://tripleten-com.github.io/simple-shop_es/'
req = requests.get(URL)
soup = BeautifulSoup(req.text, 'lxml') # Utiliza lxml como parser
name_products = []  # Lista donde se almacenan los nombres de los productos
price = []

cards = soup.find('ul',attrs={"class": "cards"}) # Encuentra el elemento <ul> con la clase "cards"

for row in cards.find_all('p', attrs={"class": "t754__title t-name t-name_md js-product-name"}): # Encuentra todos los elementos <p> con la clase "t754__title t-name t-name_md js-product-name"
    name_products.append(row.text)

for row in soup.find_all('p', attrs={"class": "t754__price-value js-product-price"}): # Encuentra todos los elementos <p> con la clase "t754__price-value js-product-price"
    price.append(row.text)

products_data = (pd.DataFrame({'name': name_products, 'price': price})) # Crea un DataFrame de pandas con los nombres y precios de los productos

print(name_products)
print(price)
print(products_data.head())
```
