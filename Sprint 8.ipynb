### Sprint 8: Solicitud GET

```python
import requests

URL = "https://practicum-content.s3.us-west-1.amazonaws.com/new-markets/Data_sprint_6/web-sites/most_famous_shipwrecks_es_1.html?etag=87662559becdeb134637e5670502d6d6"

req = requests.get(URL)
#req.raise_for_status()

print("Status code:", req.status_code)
#print(req.text[:1000])  # Print the first 1000 characters of the response text

print(req.text)
print(req.text.count("Titanic"))  # Count occurrences of "Titanic" in the response text
```
