```python
### Carga de datos
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

data = pd.read_csv('E:/datasets/travel_insurance_us.csv')

data.head(10)
```

```python
train, valid = train_test_split(data, test_size=0.25, random_state=12345)

features_train = train.drop(['Claim'], axis=1)
features_valid = valid.drop(['Claim'], axis=1)

target_train = train['Claim']
target_valid = valid['Claim']

print(features_train.shape)
print(features_valid.shape)
print()
print(len(train)+len(valid) == len(data))
print(len(valid) == len(data)*0.25)
```

```python
### Entrenamiento de pilotos
### Entrenar modelo usando datos sin procesar

model = LogisticRegression()
model.fit(features_train, target_train)
```
